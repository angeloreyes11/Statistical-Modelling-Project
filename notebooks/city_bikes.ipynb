# CityBikes

Send a request to CityBikes for the city of your choice. 

```python
import requests

city_name = 'denver'
api_endpoint = 'https://api.citybik.es/v2/networks'
response = requests.get(api_endpoint)

if response.status_code == 200:
    data = response.json()
    denver_network = None
    
    for network in data['networks']:
        if city_name.lower() in network['location']['city'].lower():
            denver_network = network
            break
    
    if denver_network:
        denver_endpoint = denver_network['href']
        denver_response = requests.get(f'https://api.citybik.es{denver_endpoint}')
        
        if denver_response.status_code == 200:
            denver_data = denver_response.json()
            print(denver_data)
        else:
            print(f"Failed to fetch Denver data. Status code: {denver_response.status_code}")
    else:
        print(f"Denver network not found in the API response.")
else:
    print(f"Failed to fetch data. Status code: {response.status_code}")

```
